```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.io import wavfile
```

```python
sample_rate, data = wavfile.read("sound-effect-thriller.wav")
print("Sample rate: ", sample_rate)
print("Data shape:", data.shape)
```

```python
if len(data.shape) > 1:
    data = data.mean(axis=1)

# Step 3: Calculate Decibel (dB) Levels
# Convert amplitude to dB, handling zero values to avoid log(0)
amplitude_in_db = 20 * np.log10(np.abs(data) + 1e-6)  # Adding a small value to avoid log(0)

log_data = np.log10(np.abs(data) + 1)

# plt.figure(figsize=(12, 6))
plt.plot( log_data)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude [dB]")
plt.title("Sound Waveform in Decibel Scale")
plt.grid(True)
plt.show()
```
